# Question 4

```python
#Create a loyalty flag using loc
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking accuracy
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['customer_id', 'loyalty_flag', 'order_number']].head(60)
```

# Question 5

```python
#Checking differences in spending habits
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

On average, new customers tend to spend the most while loyal customers spend the least in terms of product prices.

# Question 6

```python
#Create a spending flag using loc
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Checking accuracy
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['customer_id', 'spending_flag', 'average_price']].head(20)
```

# Question 7

```python
#Create a order frequency flag using loc
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
#Checking accuracy
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['customer_id', 'frequency_flag', 'median_days_since_prior_order']].head(20)
```
